## 4.1. (0.2 балла)
Как по дням недели распределяется активность студентов?

```python
import matplotlib.pyplot as plt

from hw3 import *

# Создаем словарь, в котором будут суммироваться активности по дням недели
activity = {'Пн': 0, 'Вт': 0, 'Ср': 0, 'Чт': 0, 'Пт': 0, 'Сб': 0, 'Вс': 0}

# Проходим по всем сообщениям и добавляем активности по дням недели в словарь
for msg in messages:
    day = parse_time(msg[4]).strftime('%a')
    activity[day] += 1

# Создаем график
days = ['Пн', 'Вт', 'Ср', 'Чт', 'Пт', 'Сб', 'Вс']
plt.bar(days, [activity[day] for day in days])

# Добавляем подписи осей
plt.xlabel('День недели')
plt.ylabel('Активность (кол-во сообщений)')

# Отображаем график
plt.show()
```

## 4.9. (0.2 балла)
Кто находится на первых 10 местах в рейтинге студентов?

```python
from hw3 import *
import random

# Создаем словарь, в котором будут суммироваться активности по правильным решениям
activity = {}

# Проходим по всем сообщениям и добавляем активности по дням недели в словарь
for msg in statuses:
    group = msg[2]
    variant = msg[1]
    person = f"{group}\nвар. {variant}"
    achievements = msg[5]
    if not activity.get(person):
        activity[person] = 0
    if type(achievements) is str:
        achievements = eval(achievements)
    activity[person] += len(achievements)


def return_value(val):
    return val[1]


plt.figure(figsize=(15, 10))

activity = dict(sorted(activity.items(), key=return_value, reverse=True))

persons = list(activity.keys())[:10]

# Генерируем список случайных цветов
colors = ['gold', '#d5d5d5', '#8C7853', '#006056']
colors_of_bars = ['gold']
current_color = 0
for i in range(1, len(persons)):
    if activity[persons[i]] == activity[persons[i-1]]:
        colors_of_bars.append(colors_of_bars[-1])
    else:
        current_color += 1
        if len(colors) > current_color:
            colors_of_bars.append(colors[current_color % len(colors)])
        else:
            colors_of_bars.append(colors[-1])

# Создаем график
plt.bar(persons, [activity[person] for person in persons], color=colors_of_bars)

# Добавляем числа над столбцами
for person in persons:
    plt.annotate(str(activity[person]), xy=(person, activity[person]), ha='center', va='bottom')

# Добавляем подписи осей
plt.xlabel('Студент')
plt.ylabel('Кол-во достижений')

plt.show()
```
